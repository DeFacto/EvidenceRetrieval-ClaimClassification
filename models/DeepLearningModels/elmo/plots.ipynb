```python
import seaborn as sns
import pandas as pd
sns.set(style="whitegrid")
```

```python
data_dict = {}

data_dict["method"] = ["Tf-idf", "vs", "wmd", "Tf-idf", "vs", "wmd"]
data_dict["sent_retrieval"] = [0.9, 0.5,0.4, 0.9, 0.5,0.4]
data_dict["claim_cls"] = [0.3, 0.9,0.3, 0.9, 0.5,0.4]
data_dict["overall_accuracy"] = [0.7, 0.3, 0.88, 0.9, 0.5,0.4]

# data_dict["sent_retrieval"] = [0.9, 0.5,0.4]
# data_dict["claim_cls"] = [[0.3, 0.9,0.3], [0.3, 0.9,0.3]]
# data_dict["overall_accuracy"] = [[0.7, 0.3, 0.88], [0.5,0.3, 0.9]]

dataframe = pd.DataFrame(data=data_dict)
dataframe
```

```python
import numpy as np
import matplotlib.pyplot as plt
 
from matplotlib.pyplot import figure
figure(num=None, figsize=(10, 10), dpi=80, facecolor='w', edgecolor='k')
# set width of bar
barWidth = 0.25
 

# set height of bar
bars1 = [0.643, 0.545, 0.719, 0.578, 0.627, 0.731]
bars2 = [0.469, 0.512, 0.426, 0.559, 0.589, 0.746]
bars3 = [0.419, 0.268, 0.123, 0.368, 0.324, 0.43]
 
# Set position of bar on X axis
r1 = np.arange(len(bars1))
r2 = [x + barWidth for x in r1]
r3 = [x + barWidth for x in r2]
 
# Make the plot
plt.bar(r1, bars1, color='r', width=barWidth, edgecolor='white', label='Sentence retrieval')
plt.bar(r2, bars2, color='b', width=barWidth, edgecolor='white', label='Claim Classification')
plt.bar(r3, bars3, color='g', width=barWidth, edgecolor='white', label='Overall accuracy')
 
# Add xticks on the middle of the group bars
# plt.xlabel('group', fontweight='bold')
plt.xticks([r + barWidth for r in range(len(bars1))], ['Vector Space', 'TF-IDF', 'WMD', 'word2vec', 'ELMO', 'BERT'])
 
# Create legend & Show graphic
plt.legend()
plt.title("Different approaches evaluated on FEVER development dataset", fontweight='bold')
plt.xlabel("Approaches", fontweight='bold')
plt.ylabel("Accuracy", fontweight='bold')
plt.savefig("plot.pdf")
plt.show()
```

```python

# Draw a nested barplot to show survival for class and sex
g = sns.catplot(x="method", y="overall_accuracy", hue="overall_accuracy", data=dataframe,
                height=6, kind="bar", palette="muted")
g.despine(left=True)
g.set_ylabels("accuracy")
```
